Pour tracer une fonction mathématique, il est nécessaire de considérer, *à minima*, un certain nombre de critères si on souhaite faire un graphique de qualité scientifique
- Choisir les limites du graphiques
- Mettre un **titre**
- Ecrire une **légende** génrale et pour chaque axe.

<div class="warn">
Ces critères doivent être réfléchis à l'avance et ne sont en aucun cas automatisé par Python

```python
# On n'oublie pas d'importer les modules nécessaires 
from pylab import *  
```

### 1.1. Création d'un jeu de données x

```python
# on génère des valeurs pour x (ici 500 valeurs de x comprises entre 0 et 10*pi)
x = linspace(0, 10*pi, 500) # on utilise "pi" défini dans la bibliothèque numpy
```

### 1.2. Graphe simple d'une fonction mathématique
<div class="warn">
    Le point-virgule en fin de ligne permet d'éviter l'affichage de résultats parasites.
    </div>

```python
# le plus simple

plot(x,cos(x))
```

```python
# En ajustant les bornes inférieures et supérieures pour chaque axe et en définissant l'ecart entre deux graduations principales

xlim(0,2*pi)                              # Défintion de la limite du tracé selon X (abscisse) entre 0 et 2*pi (abscisse)
ylim(-1, 1)                               # Défintion de la limite du tracé selon Y (ordonnée) entre -1 et 1

xticks(arange(0, 2*pi, step=1.0))         # Graduation principale selon X tous les 1
yticks(arange(-1, 1.1, step=0.2))         # Graduation principale selon Y tous les 0.2

axhline(y=.0,color='black',linewidth=0.5) # ajout d'une ligne horizontale à x=0
axvline(x=pi,color='black',linewidth=0.5) # ajout d'une ligne verticale à y= pi

text(2, 0.4, 'cos(x)')                    # ajout de texte sur une figure

plot(x,cos(x))
```

```python
# En définissant l'ecart entre deux graduations principales et en rajoutant une graduation secondaire

xlim(0, 2*pi)
ylim(-1, 1)

axhline(y=0.0,color='black',linewidth=0.5)
axvline(x=pi,color='black',linewidth=0.5)

minorticks_on()
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))     # Graduation principale selon X tous les 1
ax.xaxis.set_minor_locator(MultipleLocator(0.2))   # Graduation secondaire selon X tous les 0.2
ax.yaxis.set_major_locator(MultipleLocator(0.5))   # Graduation principale selon Y tous les 0.5
ax.yaxis.set_minor_locator(MultipleLocator(0.1))   # Graduation secondaire selon Y tous les 0.1

plot(x,cos(x)) 
```

```python
# En ajoutant un titre (taille = 20, couleur rouge, italique et gras)

xlim(0, 2*pi)
ylim(-1, 1)

axhline(y=.0,color='black',linewidth=0.5)
axvline(x=pi,color='black',linewidth=0.5)

minorticks_on()
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))  
ax.xaxis.set_minor_locator(MultipleLocator(0.2))   
ax.yaxis.set_major_locator(MultipleLocator(0.5)) 
ax.yaxis.set_minor_locator(MultipleLocator(0.1))  

title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')   # Ajout d'un titre

plot(x,cos(x)) 
```

```python
# En ajoutant une légende 

xlim(0, 2*pi)
ylim(-1, 1)

axhline(y=.0,color='black',linewidth=0.5)
axvline(x=pi,color='black',linewidth=0.5)

minorticks_on()
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))  
ax.xaxis.set_minor_locator(MultipleLocator(0.2))   
ax.yaxis.set_major_locator(MultipleLocator(0.5)) 
ax.yaxis.set_minor_locator(MultipleLocator(0.1))  

title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')  

plot(x, cos(x), label="cos(x)") # Tracé avec la définition de la légende
legend()                        # Affichage de la légende
```

NB : la légende sera placée automatiquement là où elle semble gêner le moins. Ici il y avait plusieurs possibilités et une a été choisie arbitrairement. Vous pouvez spécifier manuellement l'endroit ou vous voulez la légende grâce à l'option `loc=` aux expressions suivantes : 'upper left', 'upper center', 'upper right', 'center left', 'center', 'center right', 'lower left', 'lower center', 'lower right'. Par exemple : 

```python
# En ajoutant une légende 

xlim(0, 2*pi)
ylim(-1, 1)

axhline(y=.0,color='black',linewidth=0.5)
axvline(x=pi,color='black',linewidth=0.5)

minorticks_on()
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))  
ax.xaxis.set_minor_locator(MultipleLocator(0.2))   
ax.yaxis.set_major_locator(MultipleLocator(0.5)) 
ax.yaxis.set_minor_locator(MultipleLocator(0.1))  

title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')  

plot(x, cos(x), label="cos(x)") # Tracé avec la définition de la légende
legend(loc='upper center')                        # Affichage de la légende en haut au centre
```

```python
# En ajoutant une légende sur les axes et en ajustant la taille de la figure

figure(figsize=(10,4))                 # Définition de la taille de la figure (à mettre en premier)

xlim(0, 2*pi)
ylim(-1, 1)

axhline(y=.0,color='black',linewidth=0.5)
axvline(x=pi,color='black',linewidth=0.5)

minorticks_on()
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))  
ax.xaxis.set_minor_locator(MultipleLocator(0.2))   
ax.yaxis.set_major_locator(MultipleLocator(0.5)) 
ax.yaxis.set_minor_locator(MultipleLocator(0.1))  

title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')  

plot(x, cos(x), label="cos(x)") 
legend()                       

xlabel("abscisses")                       # Définition et tracé de la légende sur l'axe des abscisses
ylabel("ordonnées")                       # Définition et tracé de la légende sur l'axe des ordonnées
```

```python
# En ajoutant une grille selon l'axe x

figure(figsize=(10,4))               

xlim(0, 2*pi)
ylim(-1, 1)

axhline(y=.0,color='black',linewidth=0.5)
axvline(x=pi,color='black',linewidth=0.5)

minorticks_on()
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(1))  
ax.xaxis.set_minor_locator(MultipleLocator(0.2))   
ax.yaxis.set_major_locator(MultipleLocator(0.5)) 
ax.yaxis.set_minor_locator(MultipleLocator(0.1))  

title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')  

plot(x, cos(x), label="cos(x)") 
legend()                      

xlabel("abscisses")                       
ylabel("ordonnées")                      

grid(axis='x', which='major', color='black', linewidth=1.0)   # pour ajouter une grille selon y, on utilise grid(axis='y') et sur les deux axes, grid(axis='both')
grid(axis='x', which='minor', color='lightgray', linewidth=0.5)   # pour ajouter une grille selon y, on utilise grid(axis='y') et sur les deux axes, grid(axis='both')
```

### 1.3. Graphe en coordonnées log  ou semi-log
Il est fréquent de représenter des données en échelle log ou semi-log. Il existe plusieurs possibilités, mais nous allons nous limiter aux fonctions `xscale()` et `yscale()`. Voici un exemple. 

```python
t = linspace(0, 10, 50)          # tableau de 50 temps entre 0 et 10 (secondes, par exemple)
C = 3 * t**2                      # tableau construit avec une puissance 

figure() 
plot( t, C, 'o')                  # trace classique

figure() 
plot( t, C, 'o') 
xscale('log')                     # axe x en echelle log 
yscale('log')                     # axe y en echelle log 
```

Notez comment les graduations sur les axes sont passées automatiquement en échelle log, et comment les labels sont passés en puissances de 10. 

Pour faire un tracé en semi-log, il suffit de n'utiliser que `xscale('log')` ou que `yscale('log')`. Exemple : 

```python
t = linspace(0, 10, 50)          # tableau de 50 temps entre 0 et 10 (secondes, par exemple)
C = 2 * exp(-3*t)                # tableau construit avec une exponentielle 

figure() 
plot( t, C, 'o')                  # trace classique

figure() 
plot( t, C, 'o') 
yscale('log')                     # axe y en echelle log, mais axe x toujours normal
```

Pour revenir à un axe normal, on peut utiliser `xscale('linear')` et `yscale('linear')`

### 1.4. Utiliser deux axes y différents sur la même figure 
Parfois on veut tracer sur la même figure deux quantités dans des unités différentes. On peut utiliser l'axe y de gauche pour une des quantités et un autre axe y à droite pour la seconde quantité. 

```python
t = linspace(0, 10, 50) 
C = 3*t**2 + 2*t - 1 
P = 6*exp(-t/10)

# Au lieu de ceci, ou on ne voit pas bien la courbe P : 
figure() 
plot(t, C, label='C')
plot(t, P, label='P')
legend() 
xlabel('t')
ylabel('C ou P')

# On peut utiliser ceci, avec l'axe de gauche pour C et l'axe de droite pour P
figure() 
ax = gca()                                  # on recupere le nom des axes par defaut 
ax.plot(t, C, label='C', color='blue')      # On trace dans ces axes 
ax.set_ylabel('C', color='blue')            # On place un label sur les axes "ax", avec la meme couleur que la courbe
ax2 = ax.twinx()                            # On cree de nouveaux axes qui partagent le meme x (donc twinx). 
                                            # Si on voulait partager un seul axe y pour deux axes x, on utiliserait twiny 
ax2.plot(t, P, label='P', color='red')      # On trace P sur les seconds axes appeles ax2
ax2.set_ylabel('P', color='red')            # On place un label sur les axes "ax2", avec la meme couleur que la courbe

xlabel('t')

# il est aussi possible de colorer les graduations si on veut... 
```
