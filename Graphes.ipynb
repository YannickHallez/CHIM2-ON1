## Représentation de données avec python : premiers pas

### Introduction et premier tracé très simple

<div class="rq">
    
Pour tracer des données, il est nécessaire de considérer, *à minima*, un certain nombre de critères si on souhaite faire un graphique de qualité scientifique
- Choisir les limites du graphiques
- Mettre un **titre**
- Ecrire une **légende** générale et pour chaque axe.

</div>
<br>
<div class="warn">
Ces critères doivent être réfléchis à l'avance et ne sont en aucun cas automatisés par Python
</div>
<br>
<div class="rq">
    
La bibliothèque graphique la plus utilisée s'appelle [**matplotlib**](https://matplotlib.org/)

<img width="450px" src="./IMAGES/Logo_Matplotlib.svg" style="margin-left:auto; margin-right:auto" id="logomatplotlib"></img>

La plupart de ses fonctions deviennent accessibles lorsqu'on importe la bibliothèque PyLab, grâce à la commande `from pylab import *`

`plot` est la plus simple des commandes de matplotlib. Elle relie des points dont les abscisses et ordonnées sont fournies en arguments. On va commencer par tracer des points y en fonction de points x, c'est-à-dire y=f(x). 

ATTENTION, la fonction plot ne fait pas de mathématiques symboliques, c'est à dire qu'elle ne "comprend" pas que vous voulez tracer une fonction mathématique. Elle va juste *placer des points sur le graphe* dont les abscisses sont données dans le tableau x et les ordonnées sont données dans le tableau  (comme Excel qui trace des courbes à l'aide de 2 colonnes). 
</div>

```python
# On n'oublie donc pas d'importer les modules nécessaires 
from pylab import *  

x = [0, 1, 2, 3, 4, 5]     # definition d'un tableau d'abscisses
y = [0, 1, 4, 9, 16, 25]   # definition d'un tableau d'ordonnées

plot(x,y)
```

<div class="warn">

Mais où sont les points ?
</div>
<br>
<div class="rq">

La commande plot trace par défaut des segments qui relient les points. 
</div>

La méthode la plus simple pour changer le type de représentation des données (lignes, symboles, etc) est de rajouter un troisième argument à la fonction plot. Cet argument est une chaine de caractères, donc entre ' ', qui peut contenir un code pour le type de ligne et/ou un code pour le type de symboles et/ou un code pour la couleur. 

Les types de lignes principaux sont:
-    \-  pour une ligne continue, 
-    -- pour des tirets, 
-    :  pour des pointillés
-    -. pour un mélange de tirets et pointillés
    
Les types de symboles principaux sont:
-    o  pour des ronds, 
-    s  pour des carrés (square), 
-    d  pour des diamants
-    v  pour des triangles vers le bas,
-    ^  pour des triangles vers le haut,
-    \*  pour des étoiles
    
Les codes pour les couleurs les plus courantes sont:
-    b  pour bleu
-    r  pour rouge
-    g  pour vert (green) 
-    k  pour noir (blacK) 
-    c  pour cyan
-    m  pour magenta 
-    y  pour jaune (yellow) 

Par défaut, python va tracer des lignes continues, pas de symboles, et une sorte de bleu clair dont le code est 'C0'. Ceci correspond donc à la chaine de caractères '-C0', avec le - pour dire "ligne continue" et le "C0" pour dire "bleu clair". Regardez les exemples ci-dessous pour définir différentes lignes de différentes couleurs. 
  

```python
# On définit un tableau de x allant de 0 à 1 avec 100 points 
x = linspace(0, 1, 100)
# On définit 4 tableaux qui vont servir d'ordonnées 
y1 =   x**2
y2 = 2*x**2
y3 = 3*x**2 
y4 = 4*x**2 
# On trace les 4 courbes avec des styles différents
plot(x, y1,  '-b') # ligne bleue
plot(x, y2, '--r') # ligne en tirets rouge
plot(x, y3,  ':g') # pointillés verts 
plot(x, y4, '-.k') # tirets-points noirs
```

Voici un autre exemple avec des symboles...

```python
# On définit un tableau de x allant de 0 à 1 avec 100 points 
x = linspace(0, 1, 20)
# On définit 4 tableaux qui vont servir d'ordonnées 
y1 =   x**2
y2 = 2*x**2
y3 = 3*x**2 
y4 = 4*x**2 
# On trace les 4 courbes avec des styles différents
plot(x, y1, 'ob') # ronds bleus
plot(x, y2, 'sr') # carrés rouges
plot(x, y3, '^g') # triangles verts 
plot(x, y4, '*k') # étoiles noires (et voilà, là vous avez la musique dans la tête, non ?  )
```

Parfois on ne peut psa faire ce qu'on voudrait avec cette syntaxe condensée. Il faut alors donner à plot les styles que l'on souhaite en utilisant des mots-clés : 
- linestyle= pour le style de ligne
- marker= pour le style de symbole 
- color= pour la couleur

Il y a d'autres mots-clé que vous pourrez trouver dans la doc de la fonction plot. Voici un petit exemple :

```python
# On définit un tableau de x allant de 0 à 1 avec 100 points 
x = linspace(0, 1, 20)
# On définit 4 tableaux qui vont servir d'ordonnées 
y1 =   x**2
y2 = 2*x**2
# On trace les 4 courbes avec des styles différents
plot(x, y1, linestyle='--', color='blue') # ligne bleues
plot(x, y2, linestyle="none", marker='s', color="orange") # carrés oranges, et il faut empêcher le tracé d'une ligne avec linestyle="none"
```

### Graphe simple d'une fonction mathématique

####  Création d'un jeu de données x

<div class="rq">

On a compris sur l'exemple précédent que pour tracer y=f(x), il faut définir un tableau de valeurs de x ($[x_1 ... x_k...x_n]$) et un tableau qui va contenir les valeurs correspondantes de f(x)  ($[f(x_1) ... f(x_k)...f(x_n)]$)

Créons deux jeux de données, xA et xB, et traçons cos(xA), puis cos(xB)
</div>

```python
# on génère un tableau de valeurs pour xA (ici 20 valeurs de x comprises entre 0 et 10*pi)
xA = linspace(0, 10*pi, 20) # on utilise "pi" défini dans la bibliothèque numpy
# En voici un autre avec 500 valeurs toujours entre 0 et 10*pi
xB = linspace(0,10*pi,500)
```

#### Tracé rapide des résultats d'une fonction

```python
plot(xA,cos(xA),marker='o')  # Attention, on ne trace pas vraiment "la fonction", 
                             # on trace les points dans le tableau numpy généré dans la mémoire quand 
                             # on écrit cos(xA). 
                             # On aurait pu faire 
                             # yA = cos(xA) pour donner un nom à ce tableau, puis 
                             # plot(xA, yA) pour afficher les points d'abscisses xA et ordonnées yA
```

```python
plot(xB,cos(xB),marker='o');
```

<div class="warn">

On comprend l'intérêt d'avoir une grille de points x suffisamment importante dans l'intervalle choisi.

Et maintenant qu'on a compris comment fonctionne plot, on peut se dispenser d'afficher les points
</div>

```python
plot(xB,cos(xB));
```

#### Tracé de deux fonctions sur le même graphe

```python
plot(xB,cos(xB));
plot(xB,sin(xB));
```

#### Personnalisation facile d'un graphe (titre, taille, label des axes)

<div class="rq">

Ne perdez pas de vue que les graphiques aident à comprendre des données, et à identifier des tendances. Il est important qu'ils soient bien conçus, et que toutes les informations nécessaires à leur analyse soient présentes, tels que le **titre des axes** et leur **unité**.
</div>

##### Ajout d'un titre au graphe et ajout des labels aux axes

```python
plot(xB,cos(xB),label='cos(x)')
title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')   # Ajout d'un titre
xlabel('angle / rad')
ylabel('cos(angle)')
legend() # affichage des légendes des axes, càd xlabel et ylabel
```

<div class="warn">
    
NB : la légende a été placée automatiquement, là où elle semble gêner le moins. Ici il y avait plusieurs possibilités et une a été choisie arbitrairement par la fonction `legend()`.

On est clairement en surinformation, il ne peut pas échapper au lecteur qu'on a tracé la fonction cos(x)....
</div>

##### Changer la dimension d'une figure

<div class="rq">

matplotlib travaille par défaut en pouces (inches) et pas dans le système métrique. 1 pouce équivaut à environ 2,54 cm. 

la commande est `figure(figsize=(valeur de la largeur,valeur de la hauteur))`
</div>

```python
figure(figsize=(10,4))

plot(xB,cos(xB),label='cos(x)')
title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')   # Ajout d'un titre
xlabel('angle / rad')
ylabel('cos(angle)')
legend() # affichage des légendes des axes, càd xlabel et ylabel
```

<div class="warn">

Comment définir la taille en cm?

Analysez le code ci-dessous, qui a pour but de tracer une figure de 12 cm de large x 6 cm de haut
</div>

```python
cm = 1/2.54 #(1 inch = 2.54 cm)
figure(figsize=(12*cm,6*cm))

plot(xB,cos(xB),label='cos(x)')
title("Fonction cosinus", fontsize=20, color="red", fontstyle='italic', fontweight='bold')   # Ajout d'un titre
xlabel('angle / rad')
ylabel('cos(angle)')
legend() # affichage des légendes des axes, càd xlabel et ylabel
```

### Ne pas oublier `plt.show()`

<div class="rq">

- Sans cette commande, la figure pourrait ne pas être affichée dans un environnement Python interactif. C'est le cas si on veut afficher deux graphiques différents dans la même cellule d'un jupyter notebook
- Dans les scripts ou les programmes Python exécutés en dehors d'un environnement interactif, `plt.show()` est généralement utilisé à la fin du script pour afficher la figure avant que le programme se termine
</div>

## Graphe en coordonnées log  ou semi-log

<div class="rq">
    
Il est fréquent de représenter des données en échelle log ou semi-log. Une [échelle logarithmique](https://fr.wikipedia.org/wiki/%C3%89chelle_logarithmique) est particulièrement adaptée pour rendre compte des ordres de grandeur.

Il existe plusieurs possibilités, mais nous allons nous limiter aux fonctions `xscale()` et `yscale()`. Voici un exemple, appliqué au tracé de la fonction $C(t) = 3t^2$
</div>

```python
t = linspace(0, 10, 50)          # tableau de 50 temps entre 0 et 10 (secondes, par exemple)
C = 3 * t**2                      # tableau construit avec une puissance 

figure() 
plot( t, C, marker='o')                  # trace classique

figure() 
plot( t, C, marker='o') 
xscale('log')                     # axe x en echelle log 
yscale('log')                     # axe y en echelle log 
plt.show()
```

<div class="warn">
Notez comment les graduations sur les axes sont passées automatiquement en échelle log, et comment les labels sont passés en puissances de 10. 

Pour faire un tracé en semi-log, il suffit de n'utiliser que `xscale('log')` ou que `yscale('log')`. Exemple : 

</div>

```python
t = linspace(0, 10, 50)          # tableau de 50 temps entre 0 et 10 (secondes, par exemple)
C = 2 * exp(-3*t)                # tableau construit avec une exponentielle 

figure() 
plot( t, C, marker='o')                  # trace classique

figure() 
plot( t, C, marker='o') 
yscale('log')                     # axe y en echelle log, mais axe x toujours normal
plt.show()
```

<div class="warn">
    
Pour revenir à un axe normal, on peut utiliser `xscale('linear')` et `yscale('linear')`
</div>

## Utiliser deux axes y différents sur la même figure 


<div class="rq">
    
Parfois on veut tracer sur la même figure deux quantités dans des unités différentes. On peut utiliser l'axe y de gauche pour une des quantités et un autre axe y à droite pour la seconde quantité. 

Commençons par tracer à la même échelle deux fonctions $C(t) = 3t^2+2t-1$ et $P(t) = \exp(-t/10)$

</div>

```python
t = linspace(0, 10, 50) 
C = 3*t**2 + 2*t - 1 
P = 6*exp(-t/10)

plot(t, C, label='C');
plot(t, P, label='P');
legend();
xlabel('t');
ylabel('C ou P');

plt.show()
```


<div class="warn">
    
La fonction P, qui est une exponentielle décroissante, apparaît à cette échelle comme une ligne horizontale.


On va donc tracer P selon une échelle adaptée. Ca revient à utiliser l'axe des ordonnées de droite. La fonction `twinx()` de matplotlib permet de créer un second axe y "de droite" partageant le même axe x qu'un premier axe y "de gauche" (*twin* = jumeau).

Son utilisation est un peu plus complexe, puisqu'il faut utiliser l'adressage "*orienté objet*" de matplotlib. Ca nécessite de faire d'abord appel à la fonction `gca()` qui renvoie une information (un objet) qui contient des informations sur les éléments graphiques de la figure (comme les lignes, les points, les labels, etc.). Rien ne vous interdit d'appeler cet objet `toto`, mais il est habituel de le nommer `ax`

</div>

```python
# On peut utiliser ceci, avec l'axe de gauche pour C et l'axe de droite pour P
plot(t, C, label='C', color='blue')         # On trace une figure
ylabel('C', color='blue')                   # On place un label sur l'axe y "standard", c'est-à-dire celui de gauche, avec la meme couleur que la courbe
xlabel('t')

ax = gca()                                  # on recupere les informations de la figure dans l'objet ax 
ax2 = ax.twinx()                            # on cree de nouveaux axes qui partagent le meme x (donc twinx). 
                                            # si on voulait partager un seul axe y pour deux axes x, on utiliserait twiny 
ax2.plot(t, P, label='P', color='red')      # on trace P sur les seconds axes appeles ax2
ax2.set_ylabel('P', color='red')            # On place un label sur les axes "ax2", avec la meme couleur que la courbe

plt.show()


# il est aussi possible de colorer les graduations si on veut... 
```
