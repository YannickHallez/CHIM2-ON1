### 1. Importation des données

Dans le contexte de l'utilisation du langage Python et de la bibliothèque "Numpy", le plus simple est d'utiliser un fichier au format "csv" qu'il est facile de générer, par exemple avec Excell ou tout autre logicile de type tableur (lien vers la page wikipedia:  [CSV](https://fr.wikipedia.org/wiki/Comma-separated_values)).

Le fichier utilisé dans cette partie est constitué de notes obtenues à un module d'enseignement, chaque étudiant ayant obtenu 3 notes.

Le fichier est donc constitué de 5 colonnes: 
colonne 1: Numéro de l'étudiant, colonne 2 à 4: Notes obtenues, colonne 5: Moyenne finale

**Extrait du fichier:**



```python
# Attention la numérotation des colonnes commence à 0 et non à 1
# Dans le cas présent, le fichier est localisé au même endroit que le fichier "python". 

Fichier='Notes.csv'
data = recfromtxt(Fichier, delimiter=";", dtype=None, usecols=(0,1,2,3,4), skip_header=1) 
print(data[:5])
```

```python
# Si le fichier ne se situe pas au même endroit, il faut indiquer le "chemin" entier d'accès au fichier.
# Ces 3 lignes de codes ne servent que d'exemple et non aucun impact sur la suite.

Fichier1='C:/Users/Franck/TRAVAIL/ENSEIGNEMENTS/ENSEIGNEMENT-2022-2023/L1-PYTHON/Notes.csv'
data1 = recfromtxt(Fichier1, delimiter=";", dtype=None, usecols=(0,1,2,3,4), skip_header=1) 
print(data1[:5])
```

```python
# Tracé de la moyenne finale (5ème colonne) en fonction du numéro étudiant (1ère colonne)
fig = figure(figsize=(15, 5), dpi=200)

title('Ensemble de Notes')
xlabel('Etudiant')
ylabel('Moyenne Finale')

xlim(1, 200)
ylim(0, 20)
grid(True)
ax = gca()
ax.yaxis.set_major_locator(MultipleLocator(10)) 

plot(data[:,0], data[:,4], '-o', color='black')

figure_name = 'resultats.png'
savefig(figure_name, format='png')     # permet de sauvegarder la figure dans un fichier au format "png"
show();
```

### Une autre manière, plus judicieuse, de représenter les notes: Des histogrammes.

```python
#Lecture du fichier en ommettant la 1ère colonne

data = recfromtxt('Notes.csv', delimiter=";", dtype=None, usecols=(1,2,3,4), skip_header=1) 

fig = figure(figsize=(15, 5), dpi=100)

# definitions du découpage des données.
bins = 10          # Bins = 10 => nombre de notes entre 0 et 2, puis entre 2 et 4, ... entre 18 et 20

hist(data, bins=bins, histtype='bar', align='mid', label=['CP', 'CT', 'TP','Final'], color=['pink','red','green','black'])

# Titre et Légendes
title('Histogramme de Notes')
legend(loc='upper right')
xlabel('Notes')
ylabel('Probabilité')
legend(loc='upper right')

# Eleménts du tracé
xlim(0, 20)
ylim(0, 70)
grid(True)
ax = gca()
ax.xaxis.set_major_locator(MultipleLocator(20/bins))  
ax.yaxis.set_major_locator(MultipleLocator(10))  

figure_name = 'histogramme.jpeg'
savefig(figure_name, format='jpeg')      # permet de sauvegarder la figure dans un fichier au format "png"
show();
```
