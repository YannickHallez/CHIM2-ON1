## Description
In a batch reactor (tank) we follow the chemical transformation of the species $A$ into the species $B$ according to the following total reaction:
\begin{equation}
    A \xrightarrow{k_1} B,
\end{equation}
and $B$ also decomposes into a species $C$ according to the total reaction:
\begin{equation}
    B \xrightarrow{k_2} C.
\end{equation}
$k_1$ = 0.005 s$^{-1}$ and $k_2$ = 0.0025 s$^{-1}$ are the rate constants of these reactions under the experimental conditions.

**We want:**
+ calculate numerically the temporal evolution of the concentrations of $A$, $B$ and $C$: $c_A(t)$, $c_B(t)$, $c_C(t)$
+ determine after what reaction time $t_{max}$ the concentration of $B$ is maximum ${c_B}_{max}$.

Using material balances and expressions of speed laws (assuming simple mechanisms), we establish the following coupled relationships:

$\frac{d c_A}{dt} = -k_1 c_A$

$\frac{d c_B}{dt} = k_1 c_A - k_2 c_B$

$\frac{d c_C}{dt} = k_2 c_B$

associated with the initial conditions:

$c_A(0)$ = 1 mM

$c_B(0)$ = 0

$c_C(0)$ = 0

**You will calculate numerically $c_A(t)$, $c_B(t)$ and $c_C(t)$ on the time interval [0, 2000 s])**


## ODE system
You will have to use the odeint function of the scipy.integrate module. This function integrates ODEs of the form:
\begin{equation}
    \frac{dY}{dt} = F(Y,t).
\end{equation}


$Y$ can be a vector, that is to say contain several elements. In other words, the previous equation can correspond to several coupled ODEs and **this is precisely the case here**.

In fact the system of equations to be solved can be put in the form:

\begin{align}
     \frac{d}{dt} & \begin{bmatrix}  
                        c_A \\            
                        c_B \\            
                        c_C               
                     \end{bmatrix} = 
                     \begin{bmatrix}
                         -k_1 c_A \\
                         k_1 c_A - k_2 c_B \\
                         k_2 c_B
                     \end{bmatrix}.
\end{align}

By defining the vector $Y$ as:
\begin{align}
     Y & = \begin{bmatrix}  
                        c_A \\            
                        c_B \\            
                        c_C               
                     \end{bmatrix}
\end{align}
We can rewrite the system of 3 equations in the form of a single equation:
\begin{equation}
    \frac{dY}{dt} = F(Y,t),
\end{equation}
where $F$ is also a vector given by:
\begin{align}
     F & = \begin{bmatrix}  
                        -k_1 Y(1) \\            
                        k_1 Y(1) - k_2 Y(2) \\            
                        k_2 Y(2)               
                     \end{bmatrix},
\end{align}
with $Y(1) = c_A$, $Y(2) = c_B$ and $Y(3) = c_C$.

The odeint function can integrate such an ODE system, it can be called once to resolve everything. However, unlike cases where there is a single equation to solve, it will be necessary to correctly pass as an argument, to odeint, **the function describing the ODE system** and **the initial conditions**.
